# 1. Local approach of use case (implementation on local environment)

## 1.1 Parse ontology

```python
onto_graph = Graph().parse(owl_file_path, format='xml')
```

## 1.2 Parse use case data

```python
use_case_data_graph = Graph().parse(data = use_case_data , format = 'turtle')
```

## 1.3 Import SPARQL rules

```python
use_case_data_graph.update(manufacturing_rules_SPARQL)
```

## 1.4 Parse SHACL rules

```python
shape_graph = Graph().parse(data = manufacturing_rules_SHACL, format='turtle')
```

## 1.5 Validate against the SHACL rules

```python
local_validation_result = validate(data_graph = use_case_data_graph, shacl_graph = shape_graph, ont_graph = onto_graph, inference='rdfs', )
conforms, results_graph, results_text = local_validation_result
```

```python
local_validation_result
```

```python
# Transform validation result to a string
local_validation_result = str(local_validation_result)

# Define prefix and suffix to find in validation report
prefix_I = 'Source Shape:'
suffix_I = '\\n'
prefix_II = 'Message:'
suffix_II = '\\n'

# Find all prefix and suffix occurences
local_occurences_start_I = [m.end() for m in re.finditer(prefix_I, local_validation_result)] # list contains all indices of prefix_I occurences
local_occurences_end_I = [local_validation_result[pos_I:].find(suffix_I) + pos_I for pos_I in local_occurences_start_I] # list contains all indices of suffix_I occurences
local_occurences_start_II = [n.end() for n in re.finditer(prefix_II, local_validation_result)] # list contains all indices of prefix_II occurences
local_occurences_end_II = [local_validation_result[pos_II:].find(suffix_II) + pos_II for pos_II in local_occurences_start_II] # list contains all indices of suffix_II occurences

# Extract and merge strings to build a list of feedback sentences
local_occurences = [local_validation_result[s_I:e_I].strip() + ': ' + local_validation_result[s_II:e_II].strip() for s_I, e_I, s_II, e_II in zip(local_occurences_start_I, local_occurences_end_I, local_occurences_start_II, local_occurences_end_II)] # list constains all RuleShape names and messages
local_occurences = [o.replace('http://www.semanticweb.org/maic#', '') for o in local_occurences] # remove namespace from list of messages
local_occurences
```

# 2. Webbased approch of use case (implementation on Stardog database)

## 2.1 Specify Stardog connection details
Set the variables below to your computer and database settings.

```python
stardog_dir = '/Users/florianschmid/STARDOG_HOME'
connection_details = {
  'endpoint': 'http://localhost:5820',
  'username': 'admin',
  'password': 'admin'
}
```

## 2.2 Create a new database in Stardog

```python
cwd = os.getcwd()
os.chdir(stardog_dir)
os.system ('stardog-admin server stop')
assert os.system ('stardog-admin server start') == 0
os.chdir(cwd)
```

```python
database_name = 'maic_database'

with stardog.Admin(**connection_details) as admin:
    if database_name in [db.name for db in admin.databases()]:
        admin.database(database_name).drop()
    db = admin.new_database(database_name)
```

## 2.3 Connect to the Stardog database and import MAIC ontology as a schema

```python
conn = stardog.Connection(database_name, **connection_details)
conn.begin()
conn.add(stardog.content.File(owl_file_path))
conn.commit()
```

## 2.4 Import use case data

```python
conn.begin()
conn.add(stardog.content.Raw(use_case_data , content_type = 'text/turtle'))
conn.commit()
```

## 2.5 Import SPARQL rules

```python
conn.begin()
conn.update(manufacturing_rules_SPARQL)
conn.commit()
```

## 2.6 Validate the database against the SHACL rules

```python
icv = stardog.connection.Connection.icv(conn)
webbased_validation_result = icv.explain_violations(stardog.content.Raw(manufacturing_rules_SHACL, content_type = 'text/turtle'))
```

```python
webbased_validation_result
```

```python
webbased_validation_result = str(webbased_validation_result)
prefix = "<binding name='message'><literal>"
suffix = '</literal></binding>'
webbased_occurences_start = [m.end() for m in re.finditer(prefix, webbased_validation_result)] # list contains all indices prefix occurences
webbased_occurences_end = [webbased_validation_result[pos:].find(suffix) + pos for pos in webbased_occurences_start] # list contains all indices suffix occurences
webbased_occurences = [webbased_validation_result[s:e].strip() for s, e in zip(webbased_occurences_start, webbased_occurences_end)] # list constains all messages
webbased_occurences = [o.replace('http://www.semanticweb.org/maic#', '') for o in webbased_occurences] # remove namespace from list of messages
webbased_occurences
```

## 2.7 Comments to validation report

StarDogs validator conforms false for empty set SPARQL select query results like the ones specified in the SPARQL 1.1 Validation Test (https://www.w3.org/2009/sparql/docs/tests/summary.html#aggregates-agg-empty-group). Therefore, validation results contain false warnings from ComponentMaterialRuleShapeI, ComponentMaterialRuleShapeII, ComponentMaterialRuleShapeIII, NegativeShapeAndSizeRuleShape, ToleranceRuleShape. Please ignore the corresponding results.
This issue was reported to the StarDog engineers (https://community.stardog.com/t/shacl-sparql-for-empty-sparql-select-query-results/3530). A StarDog internal ticket to resolve the problem already exists.

The StarDog SHACL validation report does not conform with the W3C documentation. SPARQL variables cannot be bound to the SHACL message as defined in the documentation (https://www.w3.org/TR/shacl/#sparql-constraints-variables). This issue was reported to the StarDog engineers (https://community.stardog.com/t/shacl-sparql-mapping-of-variables-to-sh-message/3477). A corresponding feature request was created.
